```python
# Step 1 Loading and Reading Data set
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance (2).csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum().sum()
```

Step 2:- Data Cleaning / Data preprocessing / data wrangling

```python
# one hot encodeing will be done to convert categorical variables to numerical variables
data['sex']=data['sex'].apply({'male':0,'female':1}.get)
```

```python
data.head()
```

```python
data['smoker']=data['smoker'].apply({'no':0,'yes':1}.get)
data['region']=data['region'].apply({'southeast':0,'southwest':1,'northeast':2,'northwest':3}.get)
```

```python
data.head()
```

Step 3 - Divide the data into dependent (y) and independent (x) variables

```python
x=data[['age','sex','bmi','children','smoker','region']]
y=data['charges']# this is dependent variable
```

```python
x.head()
```

```python
y.head()
```

Step 4- Splitting the Data into Training and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)# the size 0.2 means 20% of data will be used for testing and remaining 80% will be used for training the ML model
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

Step 5 - creating Ml model by using linear Regression algorithim

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()# Linear regression algorithim gets activated
```

```python
regression.fit(x_train,y_train) # fit () is used to train  the ml model by providing x_train and y_train as input variables
```

```python
predictions=regression.predict(x_test)# predict () is used to predict the output from machine learning model which gets compared to y_test
```

```python
predictions# charges predicted by the ml model (predicted values)
```

```python
regression.score (x,y) # testing the accuracy of ml model
```
